```python
from rdkit import Chem
from rdkit.Chem import AllChem
import pandas as pd
import numpy as np

def smiles_to_fingerprint_matrix(smiles_list, radius=3, n_bits=2048):
    """
    Převod seznamu SMILES na binární matici Morgan Fingerprintů.
    Každý sloupec odpovídá bitu fingerprintu a každá řádka sloučenině.
    """
    fingerprints = []
    for smiles in smiles_list:
        mol = Chem.MolFromSmiles(smiles)
        if mol is not None:
            fp = AllChem.GetMorganFingerprintAsBitVect(mol, radius, nBits=n_bits)
            fingerprints.append(list(fp))  # Konverze fingerprintu na seznam bitů
        else:
            fingerprints.append([np.nan] * n_bits)  # Pokud je SMILES neplatný
    return pd.DataFrame(fingerprints, columns=[f'Bit_{i}' for i in range(n_bits)])



# Vypočítání fingerprint matice
fingerprint_matrix = smiles_to_fingerprint_matrix(df_smiles)
X = fingerprint_matrix.loc[:, (fingerprint_matrix.sum(axis=0) > 0)]
# Zobrazení výsledků
print(X)
```

```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from scipy.cluster.hierarchy import dendrogram, linkage

def h_cluster(X, y=None, methods=('euclidean',), metrics=('single'), color_thresholds=None):
    if y is not None:
        labels = set(y)
        colors = list(mcolors.BASE_COLORS.keys())[0:len(labels)]
        label_c = {str(label) : color  for label, color in zip(labels, colors)}
    else:
        label_c = None

    n_row = len(metrics)
    n_col = len(methods)
    fig, axes = plt.subplots(n_row, n_col, figsize=(15*n_col, 6*n_row))
    linkage_map = {}
    for r, metr in enumerate(metrics):
        linkage_map[metr] = {}
        for c, meth in enumerate(methods):
            ax = axes.flat[r*n_col + c]
            # The default metric is euclidean - make sure it's printed properly in the plot-title
            if metr is None:
                metr = 'euclidean'
            print("{0} : {1}, {2} : {3}".format(r,metr, c, meth))
            Z = linkage(X, method=meth, metric=metr)
            #print(Z)
            dendrogram(Z, ax=ax, labels=y if y is not None else None, color_threshold=color_thresholds[r] if color_thresholds is not None else None) # ax=axes[i,j],
            ax.set_title("{0} linkage, {1} metric".format(meth, metr))
            ax.set_ylabel('Distance')
            ax.set_xlabel('Label')
            linkage_map[metr][meth] = Z

    # Color all observations according to their labels
    if y is not None:
        for ax in axes.flat:
            xlbls = ax.get_xmajorticklabels()
            for lbl in xlbls:
                lbl.set_color(label_c[lbl.get_text()])

    return linkage_map
```

```python
from scipy.spatial.distance import minkowski

metrics = ['jaccard', lambda x,y_labels : minkowski(x,y_labels, p=1)]
color_thresholds = [0.6, 45]
methods = ['complete']
linkage_map = h_cluster(X, y_labels.to_numpy(), methods=methods, metrics=metrics, color_thresholds=color_thresholds)
```

```python
import pandas as pd
import numpy as np
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.spatial.distance import pdist, squareform
from rdkit import Chem
from rdkit.Chem import AllChem
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder

# Function to calculate Jaccard distance matrix
def calculate_jaccard_matrix(smiles_list, radius=3, n_bits=2048):
    mols = [Chem.MolFromSmiles(smiles) for smiles in smiles_list]
    fingerprints = [AllChem.GetMorganFingerprintAsBitVect(mol, radius, nBits=n_bits) for mol in mols]
    bit_vectors = np.array([list(fp) for fp in fingerprints])
    jaccard_distances = pdist(bit_vectors, metric='jaccard')
    return squareform(jaccard_distances)

# Generate synthetic data (replace with your dataset)
data = {
    "SMILES": df_smiles,
    "Cluster_Label": y_labels
}

df = pd.DataFrame(data)

fingerprint_matrix = smiles_to_fingerprint_matrix(df_smiles)
X = fingerprint_matrix.loc[:, (fingerprint_matrix.sum(axis=0) > 0)]
# Zobrazení výsledků


#jaccard_matrix = calculate_jaccard_matrix(df["SMILES"])

# Perform hierarchical clustering with average linkage (better for Jaccard distances)
#linkage_matrix = linkage(jaccard_matrix, method="complete")  # Use average linkage here
linkage_matrix = linkage(X, method='complete', metric='jaccard')
# Assign colors to clusters
unique_labels = df["Cluster_Label"].unique()
label_encoder = LabelEncoder()
encoded_labels = label_encoder.fit_transform(df["Cluster_Label"])
label_colors = {label: plt.cm.tab10(i / len(unique_labels)) for i, label in enumerate(unique_labels)}

# Mapping of encoded cluster labels to their colors
cluster_color_map = {i: label_colors[label] for i, label in enumerate(unique_labels)}

# Create the dendrogram
plt.figure(figsize=(20, 8))
dendro = dendrogram(
    linkage_matrix,
    labels=df["Cluster_Label"].values,
    leaf_rotation=0,
    leaf_font_size=10,
    color_threshold=0,  # Adjust this threshold as needed
    above_threshold_color='gray',  # Color for distant clusters
  # Color for closer clusters
)

# Manually assign colors to clusters based on the labels
for i, d in enumerate(dendro['dcoord']):
    for j in range(2):
        x = dendro['icoord'][i][j]
        y = dendro['dcoord'][i][j]
        plt.scatter(x, y, color=cluster_color_map[encoded_labels[i]], s=30)

# Add the legend for clusters
legend_patches = [plt.Line2D([0], [0], marker='o', color=color, label=label, linestyle='', markersize=10)
                  for label, color in label_colors.items()]
plt.legend(handles=legend_patches, title="Clusters")

plt.title("Hierarchical Clustering Dendrogram with Jaccard Distance")
plt.xlabel("SMILES")
plt.ylabel("Distance")
plt.show()
```
