```python
X = dataset.getFeatures(concat=True)
X
```

```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from scipy.cluster.hierarchy import dendrogram, linkage

def h_cluster(X, y=None, methods=('euclidean',), metrics=('single'), color_thresholds=None):
    if y is not None:
        labels = set(y)
        colors = list(mcolors.BASE_COLORS.keys())[0:len(labels)]
        label_c = {str(label) : color  for label, color in zip(labels, colors)}
    else:
        label_c = None

    n_row = len(metrics)
    n_col = len(methods)
    fig, axes = plt.subplots(n_row, n_col, figsize=(8*n_col, 6*n_row))
    linkage_map = {}
    for r, metr in enumerate(metrics):
        linkage_map[metr] = {}
        for c, meth in enumerate(methods):
            ax = axes.flat[r*n_col + c]
            # The default metric is euclidean - make sure it's printed properly in the plot-title
            if metr is None:
                metr = 'euclidean'
            print("{0} : {1}, {2} : {3}".format(r,metr, c, meth))
            Z = linkage(X, method=meth, metric=metr)
            dendrogram(Z, ax=ax, labels=y if y is not None else None, color_threshold=color_thresholds[r] if color_thresholds is not None else None) # ax=axes[i,j],
            ax.set_title("{0} linkage, {1} metric".format(meth, metr))
            ax.set_ylabel('Distance')
            ax.set_xlabel('Label')
            linkage_map[metr][meth] = Z

    # Color all observations according to their labels
    if y is not None:
        for ax in axes.flat:
            xlbls = ax.get_xmajorticklabels()
            for lbl in xlbls:
                lbl.set_color(label_c[lbl.get_text()])

    return linkage_map
```

```python
y = df.pchembl_value_Mean > 7.5
y
```

```python
from scipy.spatial.distance import minkowski

metrics = ['jaccard', lambda x,y : minkowski(x,y, p=1)]
color_thresholds = [0.8, 45]
methods = ['complete']
linkage_map = h_cluster(X, y.to_numpy(), methods=methods, metrics=metrics, color_thresholds=color_thresholds)
```

```python
linkage_map
```

```python
from scipy.cluster.hierarchy import fcluster
fl = fcluster(linkage_map['jaccard']['complete'], 12, criterion='maxclust')
fl
```

```python
dataset.addProperty('Cluster', [f"Cluster {i}" for i in fl])
```

```python
from scaffviz.clustering.manifold import Manifold

class Method(Manifold):

    def __init__(self, method):
        self.method = method

    @staticmethod
    def scale(X):
        return dataset.feature_standardizer(X) if dataset.feature_standardizer else X

    def fit(self, X):
        self.method.fit(self.scale(X))
        return self

    def transform(self, X):
        return self.method.fit_transform(self.scale(X))

    def fit_transform(self, X):
        return self.method.fit_transform(self.scale(X))

    def __str__(self):
        return "Embedding"

from sklearn.decomposition import PCA

# make PCA
n_components = 10
pca = PCA(n_components=n_components)

# make TSNE
from sklearn.manifold import TSNE

tsne = TSNE(perplexity=10, random_state=42)

# make UMAP
import umap

umap = umap.UMAP(n_components=n_components, n_neighbors=30, min_dist=0.25, metric='correlation', random_state=42)

method = Method(umap)
```

```python
from scaffviz.depiction.plot import Plot

splt = Plot(method)
splt.plot(
    dataset,
    recalculate=True,
    color_by="Cluster",
    interactive=True,
    viewport_height=800,
    port=9999
)
```
