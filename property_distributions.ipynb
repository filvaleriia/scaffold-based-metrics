```python
def unique_scaffolds_in_splits(type):
    mean = []
    std = []
    median = []
    for y in range(5):
        df = pd.read_csv(f"data/input_recall_sets_new/Glucocorticoid_receptor/cIS_Glucocorticoid_receptor_{type}_{y}.csv", header = None)
        df.columns = ['smiles']
        df['scaffold'] = [MurckoScaffoldSmiles(\
                                         Chem.MolToSmiles(MakeScaffoldGeneric\
                                       (Chem.MolFromSmiles(x)))) for x in df.smiles]
        df['fps'] = [AllChem.GetMorganFingerprintAsBitVect(Chem.MolFromSmiles(r), 3, nBits=2048) for r in df['scaffold'].tolist()]
        df = df.drop_duplicates(subset=['scaffold'])
        df = df.reset_index(drop=True)



        
        tanimoto_arr_1 = [DataStructs.TanimotoSimilarity(x[0],x[1]) for x in combinations(df['fps'], 2)]
        print("Diversity inside splitting, using Tanimoto similarity")
        print(f"Mean value for {type} {y}: ", np.mean(np.array(tanimoto_arr_1)), np.std(tanimoto_arr_1))
        print(f"Median value for {type} {y}: ", np.median(np.array(tanimoto_arr_1)), np.std(tanimoto_arr_1))
        print(f"Max value for {type} {y}: ", np.max(np.array(tanimoto_arr_1)), np.std(tanimoto_arr_1))
        print(f"Min value for {type} {y}: ", np.min(np.array(tanimoto_arr_1)))
        print("--------------------------------------------------------------------------------")
        mean.append(np.mean(np.array(tanimoto_arr_1)))
        median.append(np.median(np.array(tanimoto_arr_1)))
        std.append(np.std(tanimoto_arr_1))
        df['cluster'] = [f'{type}_{y}' for e in range(len(df))]
        if y == 0:
            df_0 = df
            df__0 = np.mean(np.array(tanimoto_arr_1))
        elif y == 1:
            df_1 = df
            df__1 = np.mean(np.array(tanimoto_arr_1))
        elif y == 2:
            df_2 = df
            df__2 = np.mean(np.array(tanimoto_arr_1))
        elif y == 3:
            df_3 = df
            df__3 = np.mean(np.array(tanimoto_arr_1))
        elif y == 4:
            df_4 = df
            df__4 = np.mean(np.array(tanimoto_arr_1))

    a = ['sim_0', 'sim_1', 'sim_2', 'sim_3', 'sim_4']

 
    # Plot scatter here
    plt.bar(a, mean)

 
    plt.errorbar(a, mean, yerr=std, fmt="o", color="r")
 
    plt.show()

    dff = pd.concat([df_0, df_1, df_2, df_3, df_4], ignore_index=True)
    return df_0, df_1, df_2, df_3, df_4, dff
```

```python
class MarkerHandler(HandlerBase):
    def create_artists(self, legend, tup,xdescent, ydescent,
                        width, height, fontsize,trans):
        return [plt.Line2D([width/2], [height/2.],ls="",
                       marker=tup[1],color=tup[0], transform=trans)]


def scatter_plot(Z, ix_color, label):
    """Scatter plots projected points, with individual colouring.
    """
    print(label)

    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(111)

    list_color  = []
    list_mak    = ["o","x","s","d","v"]
    list_lab    = ['cluster_0','cluster_1','cluster_2','cluster_3','cluster_4']
    count0 = 0
    count1 = 0
    count2 = 0
    count3 = 0
    count4 = 0
    
    for i in range(len(ix_color)):  
        if i%1000 == 0:
            print(i)
        if label[i] == 0:
            markers = 'o'
            if count0 == 0:
                list_color.append(colors[ix_color[i], ...][None])
                count0 += 1
        elif label[i] == 1:
            markers = 'x'
            if count1 == 0:
                list_color.append(colors[ix_color[i], ...][None])
                count1 += 1
        elif label[i] == 2:
            markers = 's'
            if count2 == 0:
                list_color.append(colors[ix_color[i], ...][None])
                count2 += 1
        elif label[i] == 3:
            markers = 'd'
            if count3 == 0:
                list_color.append(colors[ix_color[i], ...][None])
                count3 += 1
        elif label[i] == 4:
            markers = 'v'
            if count4 == 0:
                list_color.append(colors[ix_color[i], ...][None])
                count4 += 1
        
        ax.scatter(Z[i, 0], Z[i, 1], c=colors[ix_color[i], ...][None],  marker = markers)
    
    ax.legend(list(zip(list_color,list_mak)), list_lab, handler_map={tuple:MarkerHandler()}) 
    plt.show()
```

```python
# settings
n_components = 2   # PCA dimensionality
n_clusters = 5     # number of clusters to simulate
n_samples = 36    # samples per cluster
n_dimensions = 16 
```

```python
ix_color = np.array(dff_labels)
ix_color
```

```python
# set random seed
np.random.seed(0)

# for colouring points
cmap = plt.get_cmap('jet')
colors = cmap(np.linspace(0, 1.0, n_clusters))

# simulate data (as in Tipping's paper)
bit = np.random.rand(n_dimensions, n_clusters) > 0.5
for c in range(n_clusters):
    bits = np.repeat(bit[:, c][None, :], n_samples, axis=0)
    flip = np.random.rand(n_samples, n_dimensions) > 0.85  # this is the 'noise level'. lower this threshold for more noise
    bits[flip] = ~bits[flip]
    if c == 0:
        ix_color = c*np.ones(n_samples, dtype=np.int8)
    else:
        ix_color = np.concatenate((ix_color, 
            c*np.ones(n_samples, dtype=np.int8)), axis=0)
        
```
