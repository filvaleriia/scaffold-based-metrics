```python
import pandas as pd
import numpy as np
```

```python
df_KL_dis_0 = pd.read_csv("data/KL_divergence/df_KL_dis_0")


def_gen_1 = []
def_gen_2 = []
generator_name = 'Molpher'
for x in [0,1,2,3,4]:
    try:
        df_KL= pd.read_csv(f"data/KL_divergence/df_KL_dis_{x}")
        def_gen_1.append(df_KL[df_KL.generator == generator_name][['KL_OS_IS']].values.flatten()[0])
        def_gen_2.append(df_KL[df_KL.generator == generator_name][['KL_OS_RS']].values.flatten()[0])
    except:
        def_gen_1.append(0)
        def_gen_2.append(0)
a = np.mean(def_gen_1)
b = np.mean(def_gen_2)

def_gen_1.append(a)
def_gen_2.append(b)
    
df_gen = pd.read_csv(f"data/results/Glucocorticoid_receptor/csk_scaffolds/dis/{generator_name}/df_all_clusters_with_mean.csv")
df_gen['KL_OS_IS'] = def_gen_1
df_gen['KL_OS_RS'] = def_gen_2
```
